# Problem 1: Plotting a trajectory

```python
# -- Plot the trajectory of the ball --

t = np.linspace(0, 5.4, num=54, endpoint=True, retstep=False, dtype=None)
X = []
Y = []

for i in range(len(t)):
    Y.append(10 * t[i] - (0.5 * 3.71) * t[i]**2)
    X.append(2.78 * t[i])

plt.grid(True, linestyle='--', alpha=0.5, color='#e0e0e0')
plt.title('Trajectory of a Mars Ball', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Horizontal Distance (m)', fontsize=11, labelpad=10)
plt.ylabel('Verticle Distance (m)', fontsize=11, labelpad=10)
plt.plot(X, Y)
```

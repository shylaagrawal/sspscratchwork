### **Variables & Printing**

Python variables can store any type of data.

There are two main ways that Python represents numbers: `int` and `float` An `int` is an integer, i.e. a positive or negative number with no decimal component. A `float` (or "floating point number") is a number that CAN have a fractional or decimal component.
* In general, any number written with a decimal point (e.g. `0.1234, 6.0`) is a float, while a number without one (e.g. `6`, `-25`) is an int.
* You can also create floats using scientific notation. `1.52e6` represents $1,520,000$

You can use `+`, `-`, `*`, and `/` to add, subtract, multiply, and divide. Other Python math operators include
* `**`: Exponentiation. `2**4 -> 16`
* `//`: "Integer division", which divides and ignores the remainder. `16 // 3 -> 5`
* `%`: "Modulus", which returns ONLY the remainder. `16 % 3 -> 1`

To represent **text**, Python uses a variable type called "string" (or just `str`), which you can create using quotation marks. You can use 'single quotes' (`' '`) or "double quotes" (`" "`) to create strings, as long as they match.
* `'This is a string!'`
* `"This is also a string, but with double quotes!"`

The **print** function displays values or text. You can print the value of any variable by typing `print(variable)`. You can even print multiple things: `print("The value of variable is", variable)`

````python
# -- EXAMPLE CODE: Variables and printing --

# Computes the area of a triangle given a base and a height

triangle_base = 8      # This is an int
triangle_height = 4.5  # This is a float

triangle_area = 0.5 * triangle_base * triangle_height

print('Triangle area:', triangle_area)

print('Type of each variable above:')
print(8, type(8))
print(4.5, type(4.5))
print('Some text!', type('Some text!'))
````

### **True, False, and Conditionals** (a.k.a. "if statements")

When you compare two values, Python outputs a "boolean", a value that's either `True` or `False`. Example:
```
print(2 < 5)
-> True
```

You can compare values in a number of ways:
* `a == b`  Equal 
* `a != b`  Not equal
* `a < b `  Strictly less than
* `a <= b`  Less than or equal to
* `a > b `  Strictly greater than
* `a >= b`  Greater than or equal to

You can also use the key words `and`, `or`, and `not` to chain together comparisons:
```
(a > 0) and (a != 10)
```

An **if statement** will only run a block of code if a condition is `True`. If it's `False`, the code is skipped. It can be followed by an `elif` ("else if") statement or an `else` statement, which will only run if the statements above them did not:
```
if [CONDITION]:
    BLOCK OF CODE
    ...
elif [CONDITION]:
    DIFFERENT CODE
    ...
else:
    THIS CODE RUNS IF NONE OF THE CONDITIONS ABOVE WERE TRUE
    ...
```

Each `if` (or `elif` or `else`) statement MUST end with a colon `:`, and it must be followed up by a block of code that's indented.

````python
# -- EXAMPLE CODE: True/False and conditionals --

# Determines which of "a", "b", or "c" is largest

a = 3
b = 6
c = 9

if (a > b) and (a > c):
    print('a is largest')
    print('Amazing for a!')
elif (b > c):
    print('b is largest')
    print('Bully for b!')
else:
    print('c is largest')
    print('Celebrate c!')
print("(This line is not indented, so it's not part of the if statement)")
````

### **Functions**

A "function" is a snippet of code that takes in a set of **arguments**, then **returns** a result (with the keyword `return`).

To **define** a function:
```
def add_args(arg1, arg2):
    # Returns the sum of arg1 and arg2
    return arg1 + arg2
```
(When a function `returns` a value, it immediately stops running.)

To **call** a function:
```

-> 16
```

Throughout the program, when we ask you to write code, we'll almost always ask you to write a function or set of functions. They are **very** useful for keeping your code organized.

You can call functions from anywhere, even from within other functions.

````python
# -- EXAMPLE CODE: Functions --

def get_larger(a, b):
    # Given two numbers "a" and "b", returns the larger one
    if a > b:
        return a
    return b

def get_largest_of_3(a, b, c):
    # Given THREE numbers "a", "b", and "c", returns the largest one
    return get_larger(get_larger(a, b), c)

get_largest_of_3(10, 20, -2)
````

# 2) 💫 Loops 💫 

**This section covers:** "For loops", "while loops", and helpful ways to use them

In many programs, it's necessary to complete the same step many times, or even a variable number of times. Examples:
* Given a number "n", print the every number from 1 to n
* Ask a user for a password until the correct password is entered
* Run a video game until a character runs out of health

These situations require you to use a **loop**: a structure that runs the same block of code on repeat until a condition is met.

## **2.1) "For" loops**

In Python, a **for loop** runs a block of code a specific, pre-defined number of times.

Here is a for loop that prints the numbers `0, 1, 2, 3, 4`:
```
for i in range(5):
    # This loop runs once for each value of i in (0, 1, 2, 3, 4)
    print(i)
```

**Syntax of a for loop:**

* Always starts with `for ___ in ____:`

* Like an "if statement", must end with a colon `:` and must be followed by a block of indented code.

* The **range** keyword tells Python how long the loop should be. A loop with `range(5)` will repeat exactly 5 times.

* `i` is known as the "loop variable": it goes up by one each time, and we can use it to keep track of how long the loop has been running.

### **In the cell below, follow along with your instructor to write an example loop:**

````python
# -- WRITE TOGETHER: A simple for loop --

# Write a loop that prints the SQUARE of each number in the range (0, 1, 2, 3, 4)
for i in range(5):
    print(i**2)
````

## **2.2) Helpful ways to use loops**

Loops are most helpful when combined with other programming techniques.

#### A) Loops and if statements

By using an **"if statement" within a loop,** you can make special things happen for only specific values if `i`

**Example:**

````python
# -- EXAMPLE CODE: An if statement in a loop --

for i in range(6):
    if i == 1:
        print(i, 'is the loneliest number')
    elif i == 3:
        print(i, 'is a party')
    else:
        print(i)
````

<br>

#### B) Accumulator loops

Another common strategy is the **accumulator loop.** In this type of loop, you create a variable with a certain value, then update it each time the loop runs. This variable *accumulates* changes as the loop runs.

**Example:** Here's a very simple "accumulator loop" that adds the numbers from $0$ to $n$:

````python
# -- EXAMPLE CODE: An accumulator loop --

# Adds the numbers from 0 to n (not including n)
n = 10

total = 0 # Initializes the total to 0
for i in range(n):
    total += i # Updates the total on each loop

print('n =', n, '| sum =', total)
````

#### C) Nested loops

Finally, you can create **nested loops**, or loops that run inside other loops.
* This is allowed as long as the loop variables have different names! The tradition is to use `i` for the first loop, `j` for the second, `k` for the third, and so on.

The example below uses a nested loop to print every pair of numbers $(i, j)$ between 0 and 5 where $i > j$:

````python
# -- EXAMPLE CODE: Nested loops --

for i in range(5):
    for j in range(i):
        print('i =', i, ' j =', j)
    print('---')
````

## **2.3) "While" loops**

A *while loop* is a different way to write a loop. It uses the `while` keyword, and it keeps running as long as some condition is `True`:
```
while [CONDITION]:
    # CODE
```

Unlike a "for loop", a "while loop" doesn't to know how many times it will run beforehand! It will run as many times as necessary until its condition is `False`. A while loop is perfect for a situation like this:
```
while password_is_incorrect():
    request_password()
```
Or:
```
while not method_of_gauss_converged():
    improve_orbit()
```

However, also unlike a for loop, a while loop can run *indefinitely*! For instance, the code below will keep running until it's manually stopped. This situation is called an *infinite loop*:
```
# This will keep running forever!
while True:
    do_something()
```
If you're not careful, it's possible (indeed, easy) for your code to get stuck in an infinite loop. Because of that, you need to be more careful when using "while loops" than "for loops".

**Example of a while loop:**

The cell below uses a "while" loop to find the first value of $n$ such that $1 + 1/2 + 1/3 + 1/4 + ... + 1/n > 10$. (This sum is called the [*harmonic series*](https://en.wikipedia.org/wiki/Harmonic_series_(mathematics)).)

We need to use a "while" loop here because we don't know beforehand how many terms we'll need to add before we reach 10.

(Note: If you need to interrupt your code because it's taking too long to run, you can use the square "interrupt this kernel" button at the top of your Jupyter notebook.)

````python
# -- EXAMPLE CODE: A while loop --

n = 0      # Stores the current denominator
total = 0  # Stores the sum of all terms so far

while (total < 10):
    n += 1        # Increases the denominator by 1
    total += 1/n  # Adds 1/n to the total

print('The sum exceeds', 10, 'after', n, 'terms')
````

By the end of this section, our goal is to write a function that can **add two vectors together:**

def add_vectors(vec1, vec2):

However, there's an immediate problem! So far we've only seen Python variables that store single values. A vector is an object that stores *multiple* values. 

Further, vectors can be any length: an $N$-dimensional vector has $N$ different elements. We'd like our function to work in any number of dimensions. How can we handle this in Python?

**The solution: use a *list* to represent a vector.**

### **3.1) What is a list?**

In Python, a list is an object that stores a sequence of values. It's represented with square brackets. (Notably, this is why you can't use square brackets when doing math.)
```
    list_example = [10, 20, 30]
```

A number's position in this list is called it's ***index***. Like the "range" function, list indices in Python start counting from **0.** (They are "***zero-indexed***"). Here, the number 10 has index 0, 20 has index 1, and 30 has index 2.

To **access** item `i` in a list, you would write `list_example[i]`:
```
    list_example[0]
    -> 10
```



You can **modify** any item in a list like normal:
```
    list_example[0] -= 3
    
    print(list_example)
    -> [7, 20, 30]

    list_example[2] = list_example[0]
    
    print(list_example)
    -> [7, 20, 7]
```

**To get the length of a list, we can use the `len` function:**
```
    len(list_example)
```

The cell below shows a list in action:

````python
# -- EXAMPLE CODE: Creating a list and accessing its values --

# Creates the list
list_example = [10, 15, 30, 60]

print(list_example)
print('Length:', len(list_example))
print('Item 0:', list_example[0])
print('Item 3:', list_example[3])

# Changes the first item
list_example[0] = 5
print('After changing list_example[0], the list is:')
print(list_example)
````

### **3.2) Lists and loops**

We want the function `add_vectors(vec1, vec2)` to work on vectors of any size, which means we can't access the elements in each vector one-by-one, e.g. `vec1[0] + vec2[0]`, `vec1[1] + vec2[1]`, ...

In general, when dealing with lists in Python, you'll want to use a *loop* that iterates over each element in the vector. The standard way to do this is with the following code:
```
    for i in range(len(list_example)):
        # Do something with list_example[i]
```

(The expression `for i in range(len(LIST_NAME)):` is one you'll probably get sick of writing by the end of the program.)

**Example:** The code below uses a for loop to compute the sum of all items in a list. It works for lists of any length, including empty lists!
* The length of an empty list is 0, so the for loop runs 0 times: `for i in range(0)`.

````python
## -- EXAMPLE CODE: Looping over a list --

def get_total(values):
    # Sums the contents of `values`
    total = 0
    for i in range(len(values)):
        total += values[i]
    return total

values = [1, 2, 3, 10, -10, 90]
print(values)
print('This list sums to', get_total(values))
print()

empty = []
print(empty)
print('This list sums to', get_total(empty))
````

A common way to create a list in Python is to start with an empty list, `[]`, and use `append()` to add new items one-by-one.

```

### **3.4) "Slicing" a list**

Index notation (e.g. `list_example[i]`) allows you to look up one element at a time in a list. But what if you want to select multiple elements or a whole section of a list?

Using a ***slice***, you can select many elements at once. "List slicing" is a very versatile tool, and one you'll find yourself using many times! Here is a rundown of how it works:

#### **Taking a slice between two indices**
```
    list_example[start : end]
```
This command selects the set of elements starting from the index **start** and ending just before **end** (NOT including end itself!).

**Example:**
```
    list_example = [0, 10, 20, 30, 40, 50]

    list_example[1 : 4] # Selects the elements at 1, 2, and 3
    -> [10, 20, 30]
```

If you don't provide a "start" or an "end", Python will assume you mean the very start or end of the list.
```
    list_example[: 4] # Selects all elements from the beginning until index 4
    -> [0, 10, 20, 30]
    
    list_example[4 :] # Selects all elements index 4 until the end
    -> [40, 50]
```

#### **Adding a "step size"**
```
    list_example[start : end : step]
```
Optionally, you can add a step size, which will cause the slice to "skip" by `step` indices at a time. You won't always need this, but it's sometimes very helpful.

**Example:**
```
    list_example[1 : 5] # Selects every element from index 1 to 5
    -> [1, 2, 3, 4]
    
    list_example[1 : 5 : 2] # Selects EVERY OTHER element from index 1 to 5
    -> [1, 3]
```
The step size has some unexpected uses! Most notably, **to reverse a list, you can use a slice with step size -1:**
```
    list_example[ : : -1] # Prints every element in reverse order
    -> [5, 4, 3, 2, 1, 0]
```


````python
# -- EXAMPLE CODE: Slicing a list --

list_example = [0, 1, 2, 3, 4, 5]

print(list_example[ : 2]) # Selects all elements up to index 2
print(list_example[2 : ]) # Selects all elements from index 2 til the end
print(list_example[ : : -1]) # Reverses the entire list
````

### **3.5) Beware! Lists are *mutable***

**What's "mutable"?** *"Mutable"* ("MUTE-able") is a technical term for *"changeable"*. It comes from the same root word as "mutate".

**What does it mean?** Unlike simpler data types, every change made to a list is made directly to that list's location in memory.

This may sound innocent, but list mutability is an extremely common source of bugs, unexpected behaviors, and headaches. To explain why and to help you avoid those, let's take a detailed look.

When working with simple data (numbers, booleans, strings, etc.), when we set one variable equal to another, changes to the new variable don't affect the old one:

````python
# -- EXAMPLE CODE: Using non-lists... --

old_value = 4
new_value = old_value
new_value += 100

print("Old value:", old_value)
print("New value:", new_value)
````

***This is not true for lists!*** When you set one list equal to another, changes to the new list also happen to the old one, and vice versa.

````python
# -- EXAMPLE CODE: Using lists... --

old_list = [4]
new_list = old_list
new_list[0] += 100

print("Old list:", old_list)
print("New list:", new_list)
````

Why does this happen?

`old_list` and `new_list` have different names, but behind-the-scenes, they both refer to the ***exact same location*** in your computer's memory. Since lists can be potentially very large, Python doesn't create a new list in your computer's memory until you ask it to do so.

When you type `print(old_list)` or `print(new_list)`, Python looks up whatever's stored in that spot and prints it. No matter which of these variables you *try* to change, there's only one list in Python's memory to which that change can be applied.

As a result, if you aren't careful, changes to one list can change completely unrelated lists in ways you don't expect!

**The solution: To avoid this issue, make a copy of a list before making changes.**

One way to copy a list is to use the `list` function. Calling `new_list = list(old_list)` causes Python to copy all the values from `old_list` to a new spot in your computer's memory.

```
    old_list = [1,2,3]
    new_list = list(old_list)
    # Now, changes to new_list won't affect old_list (and vice versa)
```

Another way is to build the new list from scratch, starting with an empty list and `appending` new values one-by-one. Which method you use will depend on the situation.

> #### **A realistic example**
>
> Imagine you have your asteroid's current position (a list of its coordinates `[x, y, z]`). You run some code to calculate where it will be in 7 days, and you find that -- according to your code -- the asteroid hasn't moved! The current position and future position are exactly the same.
>
> You check your math backwards and forwards, but nothing seems to be wrong.
> 
> Errors like this often happen when you create a new list from an old one without copying it, like this:
>
> ```
>     future_position = current_position
>     # Changes to future_position
>     
> ```
> After running this code, `future_position` and `current_position` refer to the same list behind the scenes. They appear to be equal because every time your code changed the asteroid's future position, it changed the CURRENT position as well.
>
> You could fix this error by creating a copy of the current position, so changes to the future position won't affect it:
> ```
>     future_position = list(current_position)
>     
> ```
> However, you'll write a lot of code during the program, and it's likely that you'll fo at least once. Remember to check for this type of mistake when you're debugging your code.


In the exercises below, you'll write several functions that take in lists. Be careful that you're always making/modifying a copies, not changing the originals.

For your reference:
```
    def func_name(input_list):

        # This is fine
        new_list = list(input_list)

        # This is also okay
        new_list = []
        for i in range( len(input_list) ):
            new_list.append( input_list[i] )

        # NO: Changes to "new_list" will change "input_list" too
        new_list = input_list

        # ALSO NO: This changes input_list directly
        input_list[0] += 10
```

````python
# -- Problem 3.1: multiply_vector --

def multiply_vector(vec, mult):
    # Given a vector "vec" (list) and a multiplier "mult" (float),
    # returns a new list representing the vector "mult * vec"

    # YOUR CODE HERE
    results = []
    for i in vec:
        results.append(i*mult)
    return results

## -- Test case --

test_vec = [10, 15, 20]
test_mult = 3
expected = [30, 45, 60]
print('Expected:', expected)
print('Actual:  ', multiply_vector(test_vec, test_mult))
````

````python
# -- Problem 3.1: get_differences --

def get_differences(values):
    # Returns a list of differences between neighboring items in the list "values"
    
    # YOUR CODE HERE
    results = []
    for i in range(1, len(values)):
        results.append(values[i]-values[i-1])
    return results
## -- Test case --

test_case = [4, 5, 7, 1, 0.5]
expected = [1, 2, -6, -0.5]
print('Expected:', expected)
print('Actual:  ', get_differences(test_case))
````

You now have the tools you need! Implement the function `add_vectors`.

This function should take two lists `vec1` and `vec2` that represent vectors (you can assume these lists have the same length), then return a new list equal to their vector sum.

Your function should output a new list, and should NOT modify `vec1` or `vec2`.

All we've given you are some test cases. You'll need to write this function from scratch.

````python
# -- Problem 3.2: Add two vectors --

# YOUR CODE HERE
def add_vectors(vec1, vec2):
    results = []
    for i in range(len(vec1)):
        results.append(vec1[i]+vec2[i])
    return results

## -- Test cases --

test_vec1 = [1, 2, 0]
test_vec2 = [0, 0, 3]
expected  = [1, 2, 3]
print('Expected:', expected)
print('Actual:  ', add_vectors(test_vec1, test_vec2))
print()

test_vec1 = [13.0, -11.2, 0.4]
test_vec2 = [-1.1,   4.1, 1.9]
expected  = [11.9,  -7.1, 2.3]
print('Expected:', expected)
print('Actual:  ', add_vectors(test_vec1, test_vec2))
print()
````

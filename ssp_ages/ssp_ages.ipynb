```python
from datetime import date
import pandas as pd
import matplotlib.pyplot as plt
DAYS_IN_A_YEAR = 365.25

# EDIT THIS CONTANT
today = date(2026, 7, 13)
```

```python
ages = pd.read_csv("ages.csv", skipinitialspace=True)
ages.head()
```

```python
ages["Age (days)"] = ages.apply(lambda row: (today - date(row["Year"], row["Month"], row["Day"])).days, axis=1)
ages["Age (years)"] = ages["Age (days)"] / DAYS_IN_A_YEAR
ages = ages.sort_values(by="Age (days)")
ages
```

```python
plt.figure(figsize=(12, 7))
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', color="#C8C8C8", zorder=1)
plt.plot(ages["Name"], ages["Age (years)"], color="#D2D2D2", linestyle="-", label="Trend Line", zorder=2)
plt.scatter(ages["Name"], ages["Age (years)"], color="#060048", zorder=3)
plt.vlines(ages["Name"], ymin=min(ages["Age (years)"]-1), ymax=ages["Age (years)"],color="#060048", linewidth=1.5)
plt.ylabel("Age in years")
plt.title("Youngest to Oldest SSP Astro Members")
plt.show()
```

```python
plt.figure(figsize=(7, 10))
plt.grid(axis='x', color="#C8C8C8", zorder=1)
plt.plot(ages["Age (years)"], ages["Name"], color="#D2D2D2", linestyle="-", label="Trend Line", zorder=2)
plt.scatter(ages["Age (years)"], ages["Name"], color="#060048", zorder=3)
plt.hlines(ages["Name"], xmin=min(ages["Age (years)"] - 0.5), xmax=ages["Age (years)"], color="#080251", linewidth=1.5)
plt.xlabel("Age in years")
plt.title("Youngest to Oldest SSP Astro Members")
plt.show()
```

### **Problem 1.2:** Heron's formula

Given the lengths of a triangle's three sides, $a$, $b$, and $c$, its area can be computed with *Heron's formula*:

$$
A = \sqrt{s (s-a) (s-b) (s-c)}
$$
where $s$ is half of the triangle's perimeter:
$$
s = \frac{1}{2} (a+b+c)
$$

**In the cell below, given the three side lengths of a triangle $a,b,c$:**
* Compute $s$ and print its value
* Compute the triangle's area  and print its value

Here are some values you can use to test your formula. (Bonus: what happens when the three values do NOT make a valid triangle?)

| a | b | c | $\rightarrow$ | s | A |
| :-: | :-: | :-: | :-: | :-: | :-: |
| 3 | 4 | 5 | $\rightarrow$ | 6 | 6 |
| 5 | 10 | 13 | $\rightarrow$ | 14 | 22.44994... |
| 5 | 5 | 10 | $\rightarrow$ | 10 | 0 |

```python
import math

# Triangle's side-lengths
a = 5
b = 10
c = 13

# YOUR CODE HERE
s = 0.5 * ( a + b + c )
A = math.sqrt(s * (s - a) * (s - b) * (s - c))
print(f"{A:.2f}")
```

### **Problem 3.2:** Heron's formula, revisited

In an earlier exercise, you were asked to implement "Heron's formula" for the area of a triangle. Now, your job is to adapt this code into a function!

As a refresher:

> Given the lengths of a triangle's three sides, $a$, $b$, and $c$, its area can be computed with *Heron's formula*:
> $$A = \sqrt{s (s-a) (s-b) (s-c)}$$
> where $s$ is half of the triangle's perimeter:
> $$s = \frac{1}{2} (a+b+c)$$

In the cell below, implement the function `get_area_heron(a, b, c)`, which returns the area of a triangle with side lengths $a$, $b$, and $c$. Ensure that it works by running the `-- TEST CASES --` cell below.

```python
import math

def get_area_heron(a, b, c):
    # Returns the area of a triangle with side lengths a, b, and c using Heron's formula
    # YOUR CODE HERE
    s = 0.5 * ( a + b + c )
    A = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return A
    
```

```python
# -- TEST CASES --

print('Expected:', 6)
print('Actual:  ', get_area_heron(3, 4, 5))
print()
print('Expected:', 22.44994)
print('Actual:  ', get_area_heron(5, 10, 13))
print()
print('Expected:', 0)
print('Actual:  ', get_area_heron(5, 5, 10))
```

#### **Task 1: HMS to degrees**

Write the function `hms_to_deg`, which takes the hours, minutes, and seconds of an angle and returns the same angle in decimal degrees.

    hms_to_deg(1, 0, 0)
    -> 15

    -> 0.25

    -> 0.004166666666666667
    
    hms_to_deg(10, 20, 30)

```python
def hms_to_deg(hours, minutes, seconds):
    # Converts an angle in (hours, minutes, seconds) to decimal degrees
    # Inputs: hours (int), minutes (int), and seconds (float) of the angle
    # Output: The angle in decimal degrees

    # YOUR CODE HERE
    return (hours + minutes/60 + seconds/3600) * (360/24)

print(hms_to_deg(1, 0, 0))
print(hms_to_deg(0, 1, 0))
print(hms_to_deg(0, 0, 1))
print(hms_to_deg(10, 20, 30))
```

#### **Task 2: DMS to degrees**

Write the function `dms_to_deg`, which takes the degrees, minutes, and seconds of an angle and returns the same angle in decimal degrees.

This function must also handle *negative* angles. To handle them, the function has an extra argument called `is_pos`. This argument should be **True** if the angle is positive and **False** if it's negative. For example, the angle

    -00°31'55"

would be entered with `is_pos == False`:

    dms_to_deg(False, 0, 31, 55)

(***Suggested approach***: Get your function working for positive angles first. Once you're certain it's working, use an "if" statement to modify the result when the angle is negative.)

Before getting checked off, make sure your function works on the inputs below:

    dms_to_deg(True, 1, 0, 0)
    -> 1
    
    dms_to_deg(True, 0, 1, 0)
    -> 0.016666666666666666

    dms_to_deg(True, 0, 0, 1)
    -> 0.0002777777777777778

    dms_to_deg(False, 10, 0, 0)
    -> -10
    
    dms_to_deg(False, 10, 20, 30)
    -> -10.341666666666667


```python
def dms_to_deg(is_pos, degrees, minutes, seconds):
    # Converts an angle in (hours, minutes, seconds) to decimal degrees
    # Inputs:
    # - is_pos (bool): True if the angle is positive, False if it's negative
    # - degrees (int), minutes (int), and seconds (float) of the angle. Each of these must be positive.
    # Output: The angle in decimal degrees

    # YOUR CODE HERE
    res = degrees + minutes/60 + seconds/3600

    if (is_pos):
        return res
    else:
        return -res
    

print(dms_to_deg(True, 1, 0, 0))   
print(dms_to_deg(True, 0, 1, 0))
print(dms_to_deg(True, 0, 0, 1))
print(dms_to_deg(False, 10, 0, 0))  
print(dms_to_deg(False, 10, 20, 30))    
```
